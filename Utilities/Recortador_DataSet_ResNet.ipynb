```python
from PIL import Image
import pandas as pd
import os
import cv2
```

```python
#mg_path = '/content/drive/MyDrive/RedNeuronal_ResNet/DataSet/star/IC3521-H01_1029_593_6.jpg'
csv_path = '/content/drive/MyDrive/red_neuronalSSD/dataset_SDSS_limpio/csv/test.csv'
img_dict = '/content/drive/MyDrive/red_neuronalSSD/dataset_SDSS_limpio/test/'
output_dict = '/content/drive/MyDrive/RedNeuronal_ResNet/DataSet_SSDLimpio/dataset/'
csv_output_dict = '/content/drive/MyDrive/RedNeuronal_ResNet/DataSet_SSDLimpio/coordenadas/'
```

```python
contador = 1
width = 64
height = 64
```

```python
contador
```

```python
dfcoor = pd.read_csv(csv_path)
file_list = os.listdir(img_dict)
filename = 'image_165.jpg'

print('Procesando Imagen y Coordenadas...')
image_path = os.path.join(img_dict , filename)
img = Image.open(image_path)
df = dfcoor[dfcoor.filename == filename]
tamaño = 5
lista_clases = []
for index, row in df.iterrows():
  xmin = int(row['xmin']) - tamaño
  ymin = int(row['ymin']) - tamaño
  xmax = int(row['xmax']) + tamaño
  ymax = int(row['ymax']) + tamaño
  new_filename = f'objeto_{contador}.jpg'
  lista_clases.append((new_filename,str(row['class'])))
  cuadro_recortado = img.crop((xmin, ymin, xmax, ymax))
  cuadro_recortado = cuadro_recortado.resize((width,height))
  img_output_path = os.path.join(output_dict,new_filename)
  cuadro_recortado.save(img_output_path)
  contador += 1
df = pd.DataFrame(lista_clases)
csv_filename = filename[0:-3] + 'csv'
csv_output_path = os.path.join(csv_output_dict,csv_filename)
df.to_csv(csv_output_path, header = True , index = False)
print('Proceso completado')
```
