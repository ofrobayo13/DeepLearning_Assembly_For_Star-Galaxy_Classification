```python
from PIL import Image
import pandas as pd
import os
```

```python
# Ruta del archivo CSV
csv_file_path = '/content/drive/MyDrive/red_neuronalSSD/dataset_SDSS/csv/test_.xlsx'
directory = '/content/drive/MyDrive/red_neuronalSSD/dataset_SDSS/original_images/test'
output_path = '/content/drive/MyDrive/red_neuronalSSD/dataset_SDSS/test/'

# Obtener la lista de archivos en el directorio
file_list = os.listdir(directory)



```

```python
count = 1
name_finder = []
for file_name in file_list:
    # Comprobar si es un archivo de imagen
    if file_name.endswith('.jpg') or file_name.endswith('.jpeg') or file_name.endswith('.png'):
        # Ruta completa de la imagen
        image_path = os.path.join(directory, file_name)
        
        # Abrir la imagen
        im = Image.open(image_path)

        left = 0
        top = 0
        right = 320
        bottom = 320
        
        
        # Recorrer en dirección vertical
        for n in range(4):
            # Recorrer en dirección horizontal
            for i in range(6):
                # Recortar la imagen
                im1 = im.crop((left, top, right, bottom))
                
                # Guardar la imagen recortada
                im1.save(output_path + 'image_' + str(count) + '.jpg')
                
                name_finder.append(('image_' + str(count) + '.jpg',file_name))
                left += 320
                right += 320
                count += 1
            
            left = 0
            right = 320
            top += 320
            bottom += 320
```
